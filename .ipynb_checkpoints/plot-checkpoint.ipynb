```python
import pandas as pd
import pickle
import numpy as np
```

```python
dfs = pickle.load(open('dfs.pkl', 'rb'))
```

```python
def plot(syms):
    arrs = []
    for sym in syms:
        df = dfs[sym]
        if df is None:
            print(sym)
        else:
            arrs.append(df.累计净值.values)

    size = min(map(len, arrs))
    arr = np.vstack(list(map(lambda x: x[:size], arrs)))
    vals = (arr / arr[:, -1][:, None]).mean(axis=0)[::-1]
    pd.Series(vals).plot()
    r1, r2 = vals[-253]/vals[0], vals[-1] / vals[-252]
    r1 = r1 ** (252/(len(vals) - 252))
    r2 = r2
    print("roe before", r1, "roe lastyear", r2)
```
